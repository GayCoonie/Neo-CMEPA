# Neo-CMEPA 2.1 converter from standard IPA

Paste IPA verse into the input cell, run it, and the notebook will:

- strip IPA framing such as `/.../` and `[...]`
- preserve line breaks / stanza breaks / spacing
- convert IPA into Neo-CMEPA 2.1
- add apostrophes between syllables whether the IPA already marks syllables or not
- approximate unsupported IPA symbols to the closest Neo-CMEPA-compatible value where possible
- report approximations and anything still not handled cleanly

Current working mappings include:

- /a/ -> а
- /aː/ -> α
- /ɛ/ -> э
- /ɛː/ -> αι
- /ɪ/ -> і
- /iː/ -> í
- /ɔ/ -> o
- /ɔː/ -> ò
- /ʊ/ -> у
- /uː/ -> ú
- /eː/ -> é
- /oː/ -> ó
- /ə/ -> ă
- /w/ -> w
- /ʍ/ -> hw
- /j/ -> y
- /r/ -> ρ
- /s/ -> contextual: syllable-initial before vowel = σ; syllable-initial before consonant = с; otherwise = ß
- /θ/ -> θ
- /ð/ -> δ
- /ʃ/ -> ш
- /tʃ/ -> ч
- /dʒ/ -> џ
- /ŋ/ -> ң
- /x, ç/ -> kh
- explicit historical handling for initial kn -> k'n and gn -> g'n

The output is plain verse text, not IPA formatting.

```python
import re
import unicodedata


class NeoCMEPAConverter:
    IPA_DELIMS = {'/', '[', ']', '⟨', '⟩'}
    STRESS_MARKS = {'ˈ', 'ˌ', '‖', '¦', 'ʼ'}
    SYL_MARKS = {'.', '·', '|', '‿'}
    LENGTH_MARKS = {'ː', ':'}
    HALF_LENGTH_MARKS = {'ˑ'}
    MODIFIER_LETTERS = {'ʲ', 'ʷ', 'ʰ', 'ˠ', 'ˤ'}
    SUPRASEGMENTALS = {'˥', '˦', '˧', '˨', '˩', 'ꜛ', 'ꜜ'}
    EDGE_PUNCT = {',', '.', ';', '!', '?', '(', ')', '{', '}', '“', '”', '‘', '’', '…', '"'}

    MULTI_PATTERNS = [
        ('t͡ʃ', 't͡ʃ'), ('t͜ʃ', 't͡ʃ'), ('ʧ', 't͡ʃ'), ('tʃ', 't͡ʃ'),
        ('d͡ʒ', 'd͡ʒ'), ('d͜ʒ', 'd͡ʒ'), ('ʤ', 'd͡ʒ'), ('dʒ', 'd͡ʒ'),
        ('t͡s', 't͡s'), ('t͜s', 't͡s'), ('ʦ', 't͡s'),
        ('d͡z', 'd͡z'), ('d͜z', 'd͡z'), ('ʣ', 'd͡z')
    ]

    VOWELS = {
        'a', 'ɑ', 'ɐ', 'æ', 'ɒ',
        'e', 'ɛ', 'œ', 'ɜ', 'ɞ', 'ʌ', 'ø',
        'i', 'ɪ', 'y', 'ʏ', 'ɨ', 'ʉ',
        'o', 'ɔ', 'ɵ', 'ɤ',
        'u', 'ʊ', 'ɯ',
        'ə'
    }

    FRONT_VOWELS = {'i', 'ɪ', 'y', 'ʏ', 'ɨ', 'ʉ', 'e', 'ɛ', 'ø', 'œ'}
    BACKISH_VOWELS = {'u', 'ʊ', 'ɯ', 'o', 'ɔ', 'ɵ', 'ɤ', 'ɒ'}

    LEGAL_ONSETS_2 = {
        ('p', 'r'), ('b', 'r'), ('t', 'r'), ('d', 'r'), ('k', 'r'), ('g', 'r'), ('f', 'r'), ('v', 'r'),
        ('θ', 'r'), ('ð', 'r'), ('ʃ', 'r'), ('x', 'r'), ('h', 'r'), ('t͡ʃ', 'r'), ('d͡ʒ', 'r'),
        ('p', 'l'), ('b', 'l'), ('k', 'l'), ('g', 'l'), ('f', 'l'), ('s', 'l'),
        ('k', 'l'), ('g', 'l'),
        ('s', 'p'), ('s', 't'), ('s', 'k'), ('s', 'm'), ('s', 'n'), ('s', 'w'),
        ('k', 'w'), ('g', 'w'), ('h', 'w'), ('t', 'w'), ('d', 'w'),
        ('m', 'j'), ('n', 'j'), ('k', 'j'), ('g', 'j'), ('f', 'j'), ('v', 'j')
    }

    LEGAL_ONSETS_3 = {
        ('s', 'p', 'r'), ('s', 't', 'r'), ('s', 'k', 'r'),
        ('s', 'p', 'l'), ('s', 'k', 'w'), ('s', 't', 'w')
    }

    HISTORICAL_INITIAL_BREAKS = {('k', 'n'), ('g', 'n')}

    EXACT_CONSONANTS = {
        'p': 'p', 'b': 'b', 't': 't', 'd': 'd', 'k': 'k', 'g': 'g',
        'f': 'f', 'v': 'v', 'm': 'm', 'n': 'n', 'ŋ': 'ң', 'l': 'l', 'r': 'ρ',
        'w': 'w', 'ʍ': 'hw', 'j': 'y', 'y': 'y', 'z': 'z', 'h': 'h',
        'θ': 'θ', 'ð': 'δ', 'ʃ': 'ш',
        't͡ʃ': 'ч', 'tʃ': 'ч', 'ʧ': 'ч',
        'd͡ʒ': 'џ', 'dʒ': 'џ', 'ʤ': 'џ',
        'x': 'kh', 'ç': 'kh'
    }

    APPROX_CONSONANTS = {
        'c': ('k', "Approximated IPA 'c' to 'k'."),
        'q': ('k', "Approximated IPA 'q' to 'k'."),
        'ɡ': ('g', "Normalized IPA 'ɡ' to 'g'."),
        'ɣ': ('kh', "Approximated voiced velar fricative 'ɣ' to 'kh'."),
        'χ': ('kh', "Approximated uvular fricative 'χ' to 'kh'."),
        'ħ': ('kh', "Approximated pharyngeal fricative 'ħ' to 'kh'."),
        'ʜ': ('kh', "Approximated voiceless epiglottal/fricative-like 'ʜ' to 'kh'."),
        'ɧ': ('kh', "Approximated 'ɧ' to 'kh'."),
        'ʒ': ('ш', "Approximated 'ʒ' to 'ш'."),
        'ʂ': ('ш', "Approximated 'ʂ' to 'ш'."),
        'ɕ': ('ш', "Approximated 'ɕ' to 'ш'."),
        'ʐ': ('ш', "Approximated 'ʐ' to 'ш'."),
        'ʑ': ('ш', "Approximated 'ʑ' to 'ш'."),
        't͡s': ('ts', "Rendered affricate 't͡s' as digraph 'ts'."),
        'd͡z': ('dz', "Rendered affricate 'd͡z' as digraph 'dz'."),
        'β': ('v', "Approximated bilabial fricative 'β' to 'v'."),
        'ɸ': ('f', "Approximated bilabial fricative 'ɸ' to 'f'."),
        'ʋ': ('w', "Approximated labiodental approximant 'ʋ' to 'w'."),
        'ɦ': ('h', "Approximated breathy 'ɦ' to 'h'."),
        'ʔ': ('h', "Approximated glottal stop 'ʔ' to 'h'."),
        'ɲ': ('n', "Approximated palatal nasal 'ɲ' to 'n'."),
        'ɳ': ('n', "Approximated retroflex nasal 'ɳ' to 'n'."),
        'ɹ': ('ρ', "Approximated 'ɹ' to 'ρ'."),
        'ɾ': ('ρ', "Approximated flap 'ɾ' to 'ρ'."),
        'ɽ': ('ρ', "Approximated retroflex flap 'ɽ' to 'ρ'."),
        'ʀ': ('ρ', "Approximated uvular trill 'ʀ' to 'ρ'."),
        'ʁ': ('ρ', "Approximated uvular fricative 'ʁ' to 'ρ'."),
        'ɬ': ('с', "Approximated voiceless lateral fricative 'ɬ' to /s/ territory; rendered conservatively as 'с'."),
        'ɮ': ('z', "Approximated voiced lateral fricative 'ɮ' to 'z'."),
        'ɫ': ('l', "Approximated dark l 'ɫ' to 'l'."),
        'ʎ': ('l', "Approximated palatal lateral 'ʎ' to 'l'."),
        'ʟ': ('l', "Approximated velar lateral 'ʟ' to 'l'."),
        'ɭ': ('l', "Approximated retroflex lateral 'ɭ' to 'l'."),
        'ɓ': ('b', "Approximated implosive 'ɓ' to 'b'."),
        'ɗ': ('d', "Approximated implosive 'ɗ' to 'd'."),
        'ʄ': ('g', "Approximated implosive 'ʄ' to 'g'."),
        'ɠ': ('g', "Approximated implosive 'ɠ' to 'g'."),
        'ʛ': ('g', "Approximated implosive 'ʛ' to 'g'.")
    }

    def __init__(self):
        self.warnings = []
        self._seen = set()

    def note(self, message):
        if message not in self._seen:
            self._seen.add(message)
            self.warnings.append(message)

    def convert_text(self, text):
        self.warnings = []
        self._seen = set()
        text = unicodedata.normalize('NFC', text)
        parts = re.split(r'(\s+)', text)
        out = []
        for part in parts:
            if part == '' or part.isspace():
                out.append(part)
            else:
                out.append(self.convert_chunk(part))
        return ''.join(out), self.warnings

    def convert_chunk(self, chunk):
        if chunk == '':
            return chunk

        pieces = re.split(r'([—–-])', chunk)
        if len(pieces) > 1:
            rebuilt = []
            for p in pieces:
                if p in {'—', '–', '-'}:
                    rebuilt.append(p)
                else:
                    rebuilt.append(self.convert_chunk(p))
            return ''.join(rebuilt)

        leading = ''
        trailing = ''
        core = chunk

        while core and core[0] in self.IPA_DELIMS:
            core = core[1:]
        while core and core[-1] in self.IPA_DELIMS:
            core = core[:-1]

        while core and core[0] in self.EDGE_PUNCT:
            leading += core[0]
            core = core[1:]
        while core and core[-1] in self.EDGE_PUNCT:
            trailing = core[-1] + trailing
            core = core[:-1]

        while core and core[0] in self.IPA_DELIMS:
            core = core[1:]
        while core and core[-1] in self.IPA_DELIMS:
            core = core[:-1]

        if not core:
            return leading + trailing

        return leading + self.convert_word(core) + trailing

    def convert_word(self, word):
        tokens = self.tokenize_word(word)
        tokens = self.insert_historical_boundaries(tokens)
        units = self.tokens_to_units(tokens)
        syllables = self.syllabify(units)
        rendered = []
        for syll in syllables:
            rendered.append(self.render_syllable(syll))
        return "'".join(rendered)

    def tokenize_word(self, word):
        tokens = []
        i = 0
        while i < len(word):
            matched = False
            for raw, canon in self.MULTI_PATTERNS:
                if word.startswith(raw, i):
                    tokens.extend(self.normalize_token(canon))
                    i += len(raw)
                    matched = True
                    break
            if matched:
                continue

            ch = word[i]

            if ch in self.IPA_DELIMS or ch in self.STRESS_MARKS or ch in self.SUPRASEGMENTALS:
                if ch in self.SUPRASEGMENTALS:
                    self.note('Tone or suprasegmental marks were ignored.')
                i += 1
                continue

            if ch == "'":
                i += 1
                continue

            if ch in self.SYL_MARKS:
                tokens.append('|')
                i += 1
                continue

            if ch in self.LENGTH_MARKS:
                tokens.append(':')
                i += 1
                continue

            if ch in self.HALF_LENGTH_MARKS:
                tokens.append(':')
                self.note("Half-length mark 'ˑ' was treated as length.")
                i += 1
                continue

            j = i + 1
            while j < len(word) and unicodedata.combining(word[j]):
                j += 1
            while j < len(word) and word[j] in self.MODIFIER_LETTERS:
                j += 1

            tok = word[i:j]
            tokens.extend(self.normalize_token(tok))
            i = j

        return tokens

    def normalize_token(self, tok):
        if tok == '':
            return []

        if tok in {'ɚ', 'ɝ'}:
            self.note("R-colored schwa was rendered as 'ə' + 'r'.")
            return ['ə', 'r']

        syllabic = '̩' in tok
        nonsyllabic = '̯' in tok
        palatal = 'ʲ' in tok
        labial = 'ʷ' in tok
        aspirated = 'ʰ' in tok
        pharyngealized = 'ˤ' in tok or 'ˠ' in tok

        if aspirated:
            self.note('Aspiration marks were ignored.')
        if pharyngealized:
            self.note('Secondary articulation marks such as velarization/pharyngealization were ignored or approximated.')

        if any(unicodedata.combining(c) for c in tok if c not in {'̩', '̯'}):
            self.note('Combining IPA diacritics were stripped or approximated.')

        stripped = ''.join(c for c in tok if not unicodedata.combining(c) and c not in self.MODIFIER_LETTERS)

        if stripped == '':
            return []

        if stripped == 'ɡ':
            self.note("Normalized IPA 'ɡ' to 'g'.")
            stripped = 'g'

        if stripped == 'ʋ':
            self.note("Approximated labiodental approximant 'ʋ' to 'w'.")
            stripped = 'w'

        if nonsyllabic:
            if stripped in self.FRONT_VOWELS:
                self.note('A nonsyllabic front vowel was approximated to glide j/y.')
                return ['j']
            if stripped in self.BACKISH_VOWELS:
                self.note('A nonsyllabic back vowel was approximated to glide w.')
                return ['w']

        if syllabic:
            self.note('Syllabic consonants were rendered as schwa + consonant.')
            base = stripped
            out = ['ə', base]
            if palatal:
                out.append('j')
            if labial:
                out.append('w')
            return out

        out = [stripped]
        if palatal:
            self.note('Palatalization was approximated by adding glide j/y.')
            out.append('j')
        if labial:
            self.note('Labialization was approximated by adding glide w.')
            out.append('w')
        return out

    def insert_historical_boundaries(self, tokens):
        if not tokens:
            return tokens
        out = []
        for i, tok in enumerate(tokens):
            out.append(tok)
            if tok == '|':
                continue
            if i + 1 < len(tokens):
                nxt = tokens[i + 1]
                prev_is_boundary = (i == 0 or tokens[i - 1] == '|')
                if prev_is_boundary and (tok, nxt) in self.HISTORICAL_INITIAL_BREAKS:
                    out.append('|')
        return out

    def tokens_to_units(self, tokens):
        units = []
        for tok in tokens:
            if tok == ':':
                if units and units[-1]['vowel']:
                    units[-1]['long'] = True
                else:
                    self.note('A length mark not attached to a vowel was ignored.')
                continue
            if tok == '|':
                units.append({'token': '|', 'vowel': False, 'long': False})
                continue
            units.append({'token': tok, 'vowel': tok in self.VOWELS, 'long': False})
        return units

    def syllabify(self, units):
        if any(u['token'] == '|' for u in units):
            sylls = []
            cur = []
            for u in units:
                if u['token'] == '|':
                    if cur:
                        sylls.append(cur)
                        cur = []
                else:
                    cur.append(u)
            if cur:
                sylls.append(cur)
            return sylls if sylls else [[]]

        nuclei = []
        i = 0
        while i < len(units):
            if units[i]['vowel']:
                j = i + 1
                while j < len(units) and units[j]['vowel']:
                    j += 1
                nuclei.append((i, j))
                i = j
            else:
                i += 1

        if not nuclei:
            return [units]

        syllables = []
        onset = units[:nuclei[0][0]]

        for idx, (start, end) in enumerate(nuclei):
            current = onset + units[start:end]
            if idx < len(nuclei) - 1:
                next_start = nuclei[idx + 1][0]
                cluster = units[end:next_start]
                onset_len = self.choose_onset_len([u['token'] for u in cluster])
                coda = cluster[:-onset_len] if onset_len else cluster
                onset = cluster[-onset_len:] if onset_len else []
                current += coda
                syllables.append(current)
            else:
                current += units[end:]
                syllables.append(current)

        return syllables

    def choose_onset_len(self, cluster_tokens):
        if not cluster_tokens:
            return 0
        max_len = min(3, len(cluster_tokens))
        for n in range(max_len, 0, -1):
            suffix = tuple(cluster_tokens[-n:])
            if self.is_legal_onset(suffix):
                return n
        return 1

    def is_legal_onset(self, tokens):
        if any(t in self.VOWELS for t in tokens):
            return False
        if len(tokens) == 1:
            return True
        if len(tokens) == 2:
            return tokens in self.LEGAL_ONSETS_2
        if len(tokens) == 3:
            return tokens in self.LEGAL_ONSETS_3
        return False

    def render_syllable(self, syll):
        rendered = []
        for i, unit in enumerate(syll):
            tok = unit['token']
            if tok == 's':
                rendered.append(self.map_s_context(syll, i))
            else:
                rendered.append(self.render_unit(unit))
        return ''.join(rendered)

    def map_s_context(self, syll, idx):
        if idx == 0:
            if idx + 1 < len(syll):
                if syll[idx + 1]['vowel']:
                    return 'σ'
                return 'с'
            return 'σ'
        return 'ß'

    def render_unit(self, unit):
        tok = unit['token']
        if unit['vowel']:
            return self.map_vowel(tok, unit['long'])
        return self.map_consonant(tok)

    def map_consonant(self, tok):
        if tok in self.EXACT_CONSONANTS:
            return self.EXACT_CONSONANTS[tok]
        if tok in self.APPROX_CONSONANTS:
            rendered, msg = self.APPROX_CONSONANTS[tok]
            self.note(msg)
            return rendered
        if tok in self.VOWELS:
            return self.map_vowel(tok, False)
        if tok in {'ʘ', 'ǀ', 'ǁ', 'ǂ', 'ǃ'}:
            self.note(f"Click consonant {tok!r} is not represented in Neo-CMEPA; it was marked explicitly.")
            return f'⟦{tok}⟧'
        self.note(f"Unhandled symbol {tok!r} was marked explicitly.")
        return f'⟦{tok}⟧'

    def map_vowel(self, base, long=False):
        if base == 'ə':
            if long:
                self.note("Long schwa was reduced to short schwa 'ă'.")
            return 'ă'

        if base in {'a', 'ɑ', 'ɐ', 'æ'}:
            if long:
                if base != 'a':
                    self.note(f"Approximated long vowel {base!r} to Neo-CMEPA long open /aː/.")
                return 'α'
            if base != 'a':
                self.note(f"Approximated vowel {base!r} to Neo-CMEPA short /a/.")
            return 'а'

        if base in {'e', 'ɛ', 'œ', 'ɜ', 'ɞ', 'ʌ', 'ø'}:
            if long:
                if base in {'e', 'ø'}:
                    if base != 'e':
                        self.note(f"Approximated long vowel {base!r} to Neo-CMEPA long close /eː/.")
                    return 'é'
                if base != 'ɛ':
                    self.note(f"Approximated long vowel {base!r} to Neo-CMEPA long open /ɛː/.")
                return 'αι'
            if base != 'ɛ':
                self.note(f"Approximated vowel {base!r} to Neo-CMEPA short /ɛ/.")
            return 'э'

        if base in {'i', 'ɪ', 'y', 'ʏ', 'ɨ', 'ʉ'}:
            if long:
                if base != 'i':
                    self.note(f"Approximated long vowel {base!r} to Neo-CMEPA long close /iː/.")
                return 'í'
            if base != 'ɪ':
                self.note(f"Approximated vowel {base!r} to Neo-CMEPA short /ɪ/.")
            return 'і'

        if base in {'o', 'ɔ', 'ɵ', 'ɤ', 'ɒ'}:
            if long:
                if base in {'o', 'ɵ', 'ɤ'}:
                    if base != 'o':
                        self.note(f"Approximated long vowel {base!r} to Neo-CMEPA long close /oː/.")
                    return 'ó'
                if base != 'ɔ':
                    self.note(f"Approximated long vowel {base!r} to Neo-CMEPA long open /ɔː/.")
                return 'ò'
            if base != 'ɔ':
                self.note(f"Approximated vowel {base!r} to Neo-CMEPA short /ɔ/.")
            return 'o'

        if base in {'u', 'ʊ', 'ɯ'}:
            if long:
                if base != 'u':
                    self.note(f"Approximated long vowel {base!r} to Neo-CMEPA long close /uː/.")
                return 'ú'
            if base != 'ʊ':
                self.note(f"Approximated vowel {base!r} to Neo-CMEPA short /ʊ/.")
            return 'у'

        self.note(f"Unhandled vowel {base!r} was marked explicitly.")
        return f'⟦{base}⟧'
```

```python
# Paste your IPA verse below. Line breaks and stanza breaks are preserved.
# You can paste broad or narrow IPA. Slashes/brackets will be stripped.

IPA_TEXT = r'''
/sɛt staːr muːsə hɔrs knixt gnat ʍeːl wɔːrd rɛːd rɔːtə faːdər θaŋk/
'''

converter = NeoCMEPAConverter()
neo_cmepa_text, notes = converter.convert_text(IPA_TEXT)

print('Neo-CMEPA output:\n')
print(neo_cmepa_text)

with open('neo_cmepa_output.txt', 'w', encoding='utf-8') as f:
    f.write(neo_cmepa_text)

print('\n' + '=' * 72)
print('Notes / approximations:')
if notes:
    for i, note in enumerate(notes, 1):
        print(f'{i}. {note}')
else:
    print('No approximations were needed.')

print('\nSaved plain-text output to neo_cmepa_output.txt')

# Optional Colab download helper:
# from google.colab import files
# files.download('neo_cmepa_output.txt')
```
